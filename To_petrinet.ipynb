```python
net, initial_marking, final_marking = pm4py.read_pnml('input/order_objects.pnml')
pm4py.view_petri_net(net, initial_marking, final_marking)
```

```python
decorations = {}

# Show place names
for p in net.places:
    decorations[p] = {"label": p.name}

# Hide invisible transitions
for i, t in enumerate(net.transitions):
    if t.label is None:   # invisible transition
        decorations[t] = {
            "label": t.name
        }

print(decorations)
pm4py.vis.view_petri_net(net, im, fm, decorations=decorations)
```
